```python
#zero shot classifications on Hugging face using Transformers
#import PyTorch
#import torch
#import tensorflow as tf
import transformers
#from transformers import pipeline

#print("PyTorch version:", torch.__version__)
#print("TensorFlow version:", tf.__version__)
from transformers import pipeline

fiction_categories = ["Fiction", "NonFiction"]
"""
pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli",
                device="cpu")
"""
from transformers.utils import is_torch_available, is_tf_available

print("Torch available:", is_torch_available())
print("TF available:", is_tf_available())
pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
#let's take a look at the first known fiction book indexed [0] only the description is to be returned
books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
pipe(sequence, fiction_categories )
```

```python
#post processing to capture the prediction scores and label that was generated by the model
import numpy as np

max_index = np.argmax(pipe(sequence, fiction_categories)["scores"])
max_label = pipe(sequence, fiction_categories)["labels"][max_index]
max_label
```

```python
def generate_predictions(sequence, categories):
  predictions = pipe(sequence, categories)
  max_index = np.argmax(predictions["scores"])
  max_label = predictions["labels"][max_index]
  return max_label
```

```python
"""evaluating how well the model is doing the task by taking a sample of both fiction and nonfiction books
the classifier predicts a label and compare it to the already known label """

from tqdm import tqdm
#tqdm will help us know the progress of the prediction as the samples are about 600 which will take some time
#tqdm is a Python library used to create progress bars for loops,
#making it easy to visually monitor the execution progress of long-running operations like downloads, computations, or data processing.

actual_categories = []
predicted_categories = []

for i in tqdm(range(0, 300)):
  sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
  actual_categories.append("Fiction")
  predicted_categories.append(generate_predictions(sequence, fiction_categories))
```

```python
for i in tqdm(range(0, 300)):
  sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
  actual_categories.append("NonFiction")
  predicted_categories.append(generate_predictions(sequence, fiction_categories))
```

```python
predictions_df = pd.DataFrame({"actual_categories": actual_categories, "predicted_cateories": predicted_categories})
```

```python
predictions_df
```

```python
predictions_df["correct_prediction"] = np.where(predictions_df["actual_categories"] == predictions_df["predicted_cateories"], 1, 0)
```

```python
predictions_df["correct_prediction"].sum()/len(predictions_df)
```

```python
isbns= []
predicted_cats= []
missing_cats = books.loc[books["simple_categories"].isna(),["isbn13", "description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0, len(missing_cats))):

  sequence = missing_cats.loc[i, "description"]
  predicted_cats.append(generate_predictions(sequence, fiction_categories))
  isbns.append(missing_cats.loc[i, "isbn13"])
```

```python
missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_cats})
```

```python
missing_predicted_df
```

```python
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
```

```python
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
books = books.drop(columns=["predicted_categories"])
```
